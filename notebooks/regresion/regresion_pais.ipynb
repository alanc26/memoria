```python
## datos atenciones deis
ruta = "../../data/processed_data/total_atenciones_promedio.csv"

df_atenciones = pd.read_csv(ruta)
df_atenciones.head()
```

```python
## datos red hospitalaria deis
ruta = "../../data/processed_data/red_hospitalaria_publica.csv"

df_red = pd.read_csv(ruta)
df_red.head()
```

```python
df_red = df_red[["Nombre Comuna","Total_centros"]]
```

```python
df_red.head(30)
```

```python
# estandarizamos columnas
df_total = df_total.rename(columns={"Comuna": "Comuna"})
df_fonasa = df_fonasa.rename(columns={"COMUNA": "Comuna"})
df_atenciones = df_atenciones.rename(columns={"Comuna": "Comuna"})
df_red = df_red.rename(columns={"Nombre Comuna": "Comuna"})

for df in [df_total, df_fonasa, df_atenciones, df_red]:
    df["Comuna"] = df["Comuna"].str.strip().str.lower()
```

```python
df_total["Comuna"] = df_total["Comuna"].replace("trehuaco", "treguaco")

df_fonasa["Comuna"] = df_fonasa["Comuna"].replace({
    "cabo de hornos (ex - navarino)": "cabo de hornos",
    "ranquil": "ránquil"
})

df_atenciones["Comuna"] = df_atenciones["Comuna"].replace({
    "ranquil": "ránquil",
    "trehuaco": "treguaco"
})

df_red["Comuna"] = df_red["Comuna"].replace({
    "aisén": "aysén",
    "chepica": "chépica",
    "con con": "concón",
    "maria elena": "maría elena",
    "maria pinto": "maría pinto",
    "tome": "tomé",
    "santa maria": "santa maría",
    "ranquil": "ránquil",
    "licanten":"licantén",
    "vichuquen": "vichuquén"
})
```

```python
sets = {
    "poblacion": set(df_total["Comuna"]),
    "beneficiarios": set(df_fonasa["Comuna"]),
    "pacientes": set(df_atenciones["Comuna"]),
    "centros": set(df_red["Comuna"]),
}

# Intersección y uniones
interseccion = set.intersection(*sets.values())
union = set.union(*sets.values())

print(f"Comunas en todos los DF: {len(interseccion)}")
print(f"Comunas distintas en conjunto: {len(union)}")
```

```python
df_merged = (
    df_total
    .merge(df_fonasa, on="Comuna", how="left")
    .merge(df_atenciones, on="Comuna", how="left")
    .merge(df_red, on="Comuna", how="left")
)
```

```python
df_merged = df_merged[~df_merged["Comuna"].isin(["antártica", "primavera"])]
```

```python
import statsmodels.api as sm

df_merged['tasa_atencion'] = df_merged['promedio_pacientes'] / df_merged['poblacion_total']
df_merged['centros_por_hab'] = df_merged['Total_centros'] / df_merged['poblacion_total']

# Variables independientes
X = df_merged[['porcentaje', 'centros_por_hab', 'PORCENTAJE_EXTRANJEROS']]
X = sm.add_constant(X)  # agrega intercepto

# Variable dependiente
y = df_merged['tasa_atencion']

# Ajuste del modelo
modelo = sm.OLS(y, X).fit()
print(modelo.summary())
```

```python
from scipy.stats import zscore

# Crear indicadores
df_2 = df_merged
df_2["demanda_pob"] = df_2["promedio_pacientes"] / df_2["poblacion_total"]
df_2["demanda_capacidad"] = df_2["promedio_pacientes"] / df_2["Total_centros"]

df_2["poblacion_total_log"] = np.log(df_2["poblacion_total"])

df_2["indice_saturacion"] = zscore(df_2["demanda_pob"]) + zscore(df_2["demanda_capacidad"])


# Variables dependiente y explicativas
y = df_2["indice_saturacion"]  # o usa 'demanda_pob' si prefieres
X = df_2[["porcentaje", "centros_por_hab", "poblacion_total_log"]]

X = sm.add_constant(X)

# Ajustar modelo OLS
modelo = sm.OLS(y, X).fit()

print(modelo.summary())
```

```python
# Crear rango de porcentaje
modelo = sm.OLS(df_merged_2["tasa_atencion"], sm.add_constant(df_merged_2[["porcentaje", "porcentaje_sq", "ruralidad", "nivel_promedio", "desviacion_estandar"]])).fit()

x_vals = np.linspace(df_merged_2["porcentaje"].min(), df_merged_2["porcentaje"].max(), 100)
y_pred = (
    modelo.params["const"]
    + modelo.params["porcentaje"] * x_vals
    + modelo.params["porcentaje_sq"] * (x_vals ** 2)
)

plt.figure(figsize=(8,5))
plt.scatter(df_merged_2["porcentaje"], df_merged_2["tasa_atencion"], alpha=0.5, label="Comunas")
plt.plot(x_vals, y_pred, color="red", lw=2, label="Tendencia cuadrática ajustada")
plt.xlabel("Porcentaje de población migrante")
plt.ylabel("Tasa de atención")
plt.title("Relación entre migración y saturación sanitaria")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
