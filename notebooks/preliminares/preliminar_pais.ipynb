```python
import pandas as pd
## csv inmigracion
ruta = "../../data/processed_data/censo/censo_inmigracion/inmigracion_pais.csv"

df_inmigracion = pd.read_csv(ruta)
df_inmigracion.head()
```

```python
## csv poblacion
ruta = "../../data/processed_data/censo/censo_poblacion/poblacion_pais.csv"

df_poblacion = pd.read_csv(ruta)
df_poblacion.head()
```

```python
df_inm_data = df_inmigracion[["Código región","Región","Inmigrantes internacionales"]]
df_pob_data = df_poblacion[["Código región","Región","Población censada"]]



# Renombrar las columnas de población antes del merge
df_pob_data = df_pob_data.rename(columns={"Población censada": "poblacion_total"})
df_inm_data = df_inm_data.rename(columns={"Inmigrantes internacionales": "poblacion_migrante"})

# Hacer el merge
df_merged = df_pob_data.merge(df_inm_data, on=["Código región", "Región"])
```

```python
df_merged["porcentaje"] = round(df_merged["poblacion_migrante"] / df_merged["poblacion_total"] * 100,2)
```
