```python
import pandas as pd

# Cargar archivo .dta (CASEN)
ruta = '../data/raw_data/raw_casen.dta' 
df = pd.read_stata(ruta, convert_categoricals=False)

# Mostrar las primeras filas
df.head()
```

```python
# Guardar todas las columnas
cols = df.columns.tolist()
```

```python
# Columnas relacionadas con salud
salud_cols = [col for col in cols if 's13' in col.lower()]
print("Columnas relacionadas con salud:", salud_cols)

# Columnas relacionadas con migración o nacionalidad
mig_cols = [col for col in cols if 'r1a' in col.lower() or 'estrato'in col.lower() or 'region' in col.lower()]
print("Columnas relacionadas con migración:", mig_cols)
```

```python
# estrato contiene comuna-area-nse(nivel socio economico)
# primer valor comuna
# area puede ser 1 o 2 -> urbano o rural
# nse(nivel socio economico) -> [1,...,7] => [Bajo, medio, alto, bajo-medio,bajo-alto, bajo-medio-alto, medio-alto]

df[df['estrato'].astype(str).str.startswith("1101")].groupby('estrato').size()
```

```python
# Selección manual de columnas clave (modificar según resultados anteriores)
cols_interes = ['estrato', 'region', 'r1a', 'r1a_esp', 'r1a_esp_cod','s13', 's13_fonasa']  
subset = df[cols_interes].copy()

# Guardar como CSV
subset.to_csv('../data/procesed_data/casen_subset.csv', index=False)
```
